### Analysis_4
Find out which tags are used to describe and represent each genre, show result using word cloud <br>
Use grenre Drama, Comedy and Sci-Fi as examples.<br>
Steps:
1. For each movie, find out the genre(s) and corresponding tag word.
2. Collect genres and their tag words in dataframe


```python
import os
import pandas as pd
import numpy as np
from pandas import DataFrame as df
from datetime import datetime
import time
from wordcloud import WordCloud, STOPWORDS
from scipy.misc import imread
import matplotlib.pyplot as plt
```

```python
rootdir = os.path.expanduser('~')+'/final/data/ml-20m-dataset'
```

```python
current_dir = os.getcwd()
```

```python
movies = pd.read_csv(rootdir+'/movies.csv')
tags = pd.read_csv(rootdir+'/tags.csv')
```

```python
stopwords = set(STOPWORDS)
```

**Get all movies' genres**

```python
genres_list = []
```

```python
def get_genre(genres):
    [genres_list.append(x) for x in genres.split("|")]
```

```python
movies['genres'].apply(get_genre).head(2)
```

```python
len(set(genres_list))
```

```python
movies_1 = movies.copy()
```

```python
for x in set(genres_list):
    movies_1[x] = movies_1['genres'].apply(lambda y: x in y)
```

```python
movies_1.head(5)
```

```python
tags.ix[1]
```

```python
tags[['movieId','tag']].head()
```

```python
df_drama_tags = movies_1[movies_1['Drama']==True][['movieId']].merge(tags[['movieId','tag']], on='movieId')
df_drama_tags.head()
```

```python
cut_text = " ".join(str(e) for e in df_drama_tags['tag'].values.tolist()) 
color_mask = imread(os.path.expanduser('~')+'/final/data/alice.jpg') 
cloud = WordCloud(background_color='white',mask=color_mask,max_words=2000,max_font_size=40,stopwords=stopwords)
cloud.generate(cut_text) 
cloud.to_file(current_dir+'/plots/ana_4_drama_tags.png')
```

```python
plt.imshow(cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
cut_text = " ".join(str(e) for e in df_comedy_tags['tag'].values.tolist()) 
color_mask = imread(os.path.expanduser('~')+'/final/data/bird.png') 
cloud = WordCloud(background_color='white',mask=color_mask,max_words=2000,max_font_size=40,stopwords=stopwords)
cloud.generate(cut_text) 
cloud.to_file(current_dir+'/plots/ana_4_comedy_tags.png')
```
